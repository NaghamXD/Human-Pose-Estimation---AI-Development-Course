```python
import os
import shutil

def find_and_delete_duplicates(whole_dataset_folder, train_folder, val_folder):
    # List all images in the whole dataset folder
    whole_dataset_images = os.listdir(whole_dataset_folder)
    
    # List images in train folder
    train_images = os.listdir(os.path.join(train_folder, 'train'))
    
    # List images in val folder
    val_images = os.listdir(os.path.join(val_folder, 'val'))
    
    # Combine train and val images
    train_val_images = set(train_images) | set(val_images)
    
    # Check and delete duplicates based on first 7 characters from whole dataset folder
    for image in whole_dataset_images:
        if any(image[:7] == train_val_image[:7] for train_val_image in train_val_images):
            image_path = os.path.join(whole_dataset_folder, image)
            os.remove(image_path)
            print(f"Deleted: {image_path}")

# Example usage:
whole_dataset_folder = "C:/Users/Nagham/HRNet-Human-Pose-Estimation/data/hr-lspet1 - Copy"
train_folder = "C:/Users/Nagham/Human-Pose-Estimation---AI-Development-Course/dataset/images/"
val_folder = "C:/Users/Nagham/Human-Pose-Estimation---AI-Development-Course/dataset/images/"

find_and_delete_duplicates(whole_dataset_folder, train_folder, val_folder)
```
